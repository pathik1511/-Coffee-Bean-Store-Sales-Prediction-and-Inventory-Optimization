```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Coffee_Stores_Data.csv")
```

```python
df.head()
```

```python
df = df.drop(['CategoryLvl1Desc','CategoryLvl2Desc','CategoryLvl3Desc','ItemType'],axis=1)
```

```python
#df.sort_values(by=['Description'])
df_desc = df.Description.unique()
df_desc.sort()
df_desc = pd.DataFrame(data=df_desc)
df_desc
```

```python
df['Months'] = pd.DatetimeIndex(df['BusinessDate']).month
df['Year'] = pd.DatetimeIndex(df['BusinessDate']).year
df
```

```python
fh_18 = df.loc[(df['StoreID'] ==18)]
fh_117 = df.loc[(df['StoreID'] ==117)]
fh_332 = df.loc[(df['StoreID'] ==332)]
# 3 Store data  :
df_a=pd.concat([fh_18,fh_117,fh_332])
df_a = df_a.sort_values(by=['Description'])
```

```python
df_a = df_a[df_a['Description']>="Everything Bagel"]
```

```python
import seaborn as sns
#z = df_a.loc[df_a["Description"] == 'Everything Bagel','Months ' == 7]
#df_a.groupby(['Description'])
#df_EB = df_a[df_a['Descriptionl'] >= Everything Bagel]
g = sns.FacetGrid(df_a, col="Months", sharex=False)
g.map(sns.boxplot, 'ReceivedQuantity', 'Description')
```

```python
df_a["BusinessDate"] = pd.to_datetime(df_a["BusinessDate"])
df_a["DayOfWeek"] = df_a["BusinessDate"].dt.weekday
df_a["IsWeekend"] = df_a["BusinessDate"].dt.weekday >= 5
```

```python
df_b = df_a.pivot_table(index='Description', columns='IsWeekend', values='SoldQuantity',aggfunc='sum')
df_b.plot(kind='bar', figsize=(15, 8), color=['red', 'Green', '#e37827', '#275444'])
```

```python
df_b = df_a.pivot_table(index='StoreID', columns='Description', values='ReceivedQuantity',aggfunc='sum')
df_b.plot(kind='bar', figsize=(15, 8), color=['red', 'Green', '#e37827', '#275444'])
```
